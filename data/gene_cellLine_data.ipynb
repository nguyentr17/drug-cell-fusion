```python
# Step1: - Merge sample info with the CCLE

merge_sampleInfoXCCLE = pd.merge(df_ccle_exp, df_sample_info, on="DepMap_ID")
#merge_sampleInfoXCCLE
```

```python
# We got the cell lines from sample info
```

```python
# Step 2: Merge (Intersection) of ParsVNN cell id with CCLE Expression

merge_cell2id_X_CCLE = pd.merge(df_cell2id, merge_sampleInfoXCCLE, on = "CCLE_Name")
merge_cell2id_X_CCLE
```

```python
# Step 3: Remove the gene name Hugo number from the point where it starts with a bracket

colName = list(merge_cell2id_X_CCLE.columns)
renames = {}
for col in colName:
    renames[col] = col.upper().split(' ')[0]    
renames

merge_cell2id_X_CCLE.rename(columns = renames, inplace = True)
merge_cell2id_X_CCLE
```
